```python
import pandas as pd
diabetes = pd.read_csv('diabetes.csv')

diabetes.info()
```

```python
diabetes.head(10)
```

```python
x = diabetes[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y = diabetes['Outcome']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=350)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()

tree_model = tree_model.fit(x_train,y_train)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from sklearn.metrics import accuracy_score

y_predict = tree_model.predict(x_test)

accuracy_score = round(accuracy_score(y_predict, y_test), 3)

print('Accuracy: ',accuracy_score)
```

```python
result = tree_model.predict([[6,148,72,35,0,33.6,0.627,50]])[0]
if result == 1:
    print('Diabetes')
else:
    print('Tidak Diabetes')
```
